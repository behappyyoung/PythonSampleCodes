```python
def shortestPath(grid, k):
    m = len(grid)
    n = len(grid[0])
    trace_q = [(0, 0, k, 0)]
    visited_with = {}
    paths = []
    def checkinvisited(i, j, ck, cl):
        if (i, j, ck) in visited_with:
            if visited_with[(i, j)] <= cl:
                return False
            else:
                visited_with[(i, j)] = cl
        else:
            visited_with[(i, j)] = cl
            
        trace_q.append((i, j, ck, cl))
        
    while trace_q:
        skip = False
        current = trace_q.pop()
        i = current[0]
        j = current[1]
        ck = current[2]
        cl = current[3]
        obs = grid[i][j]
        print(current)

                
        if not skip:
            visited_with[(i, j, ck)] = cl 

            if j-1 >=0:
                if ck-grid[i][j-1] >=0:
                    checkinvisited(i, j-1, ck-grid[i][j-1], cl+1)
            if i-1>=0:
                if ck-grid[i-1][j] >=0:
                    checkinvisited(i-1, j, ck-grid[i][j-1], cl+1)
                    
            if i < m-1:
                if ck-grid[i+1][j] >=0:
                    checkinvisited(i+1, j, ck-grid[i+1][j], cl+1)
                if j < n-1:
                    if ck-grid[i][j+1] >=0:
                        checkinvisited(i, j+1, ck-grid[i][j+1], cl+1)
            else:
                if j < n-1:
                    if ck-grid[i][j+1] >=0:
                        checkinvisited(i, j+1, ck-grid[i][j+1], cl+1)
                else:
    #                 for vw in visited_with:
    #                     print(vw)
                    paths.append(cl)
    #                 return cl
    print(paths)               
    return min(paths) if paths else -1


    
```
