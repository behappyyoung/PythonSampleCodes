```python
"""

"""
from collections import defaultdict

def numBusesToDestination(routes, source, target):


    visited = set()
    s_set = set()
    t_set = set()
    routes_set = set()
    
    for r in routes:
        routes_set.add(tuple(r))
        if source in r:
            s_set = tuple(r)
        if target in r:
            t_set = tuple(r)
    
#     travel
    def travel(start, start_set, c_count):
        if target in start_set:
            return c_count
        if start_set in visited:
            return -1
        visited.add(start_set)
        total_rides = -1

        for i in start_set:
            current_min = len(routes_set) +1
            for r in routes_set:
                if i in r:
                    current_ride = travel(i, r, c_count+1)
                    if current_ride != -1:
                        current_min = min(current_ride, current_min) 
                    print(start, start_set, i, r, c_count, current_ride, current_min)

            if current_min != len(routes_set) +1:
                total_rides = current_min 
            
        return total_rides

    
    bus_rides = travel(source, s_set, 1)
     
    
    print(list(routes_set), s_set, t_set, visited)
    
    
    return bus_rides
    
routes = [[1,2,7],[3,6,7]]
source = 1
target = 6
routes = [[7,12],[4,5,15],[6],[15,19],[9,12,13]]
source = 15
target = 12
routes = [[1,2,7],[3,7],[3,4,5,6]]
source = 1
target = 6
routes = [[0,1,6,16,22,23],[14,15,24,32],[4,10,12,20,24,28,33],[1,10,11,19,27,33],[11,23,25,28],[15,20,21,23,29],[29]]
source =4
target =21
numBusesToDestination(routes, source, target)
```
