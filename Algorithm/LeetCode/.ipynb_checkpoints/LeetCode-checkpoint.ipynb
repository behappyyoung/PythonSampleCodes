```python
def isHappy(n):
        nums = set()
        while n != 1:
            if n in nums:
                return False
            nums.add(n)
            print(nums)
            sumP = 0
            while n > 0:
                sumP += (n % 10)**2
                n //= 10
            n = sumP
        return True

isHappy(2)
```

```python
def searchSub(org_a, target):
    sub_arr = []
    i = 0
    print(org_a, target)
    while i < len(org_a):
#         print(i)
        if org_a[i] == target:
            sub_arr.append(org_a[i])
            return sub_arr
        elif org_a[i] < target:
            if i == len(org_a) -1:
                return False
            
            sub_arr.append(org_a[i])
            new_target = target - org_a[i]
            
            sub_sub = searchSub(org_a[i+1:], new_target)
            if not sub_sub:
                sub_arr = []
                i += 1
            else:
                sub_arr += sub_sub
                return sub_arr
        else:
            i += 1
        
    return False
    
def matchStick(arr):
    sumofarr = sum(arr)
    if sumofarr  % 4 != 0:
        return False
    par = sumofarr//4
    rarr = sorted(arr, reverse=True)
    if rarr[0] > par:
        return False
    print(rarr, par)
    while rarr:
        if rarr[0] == par:
            rarr = rarr[1:]
        else:
            sub_list = [rarr[0]]
            sub_sub_list = searchSub(rarr[1:], par-rarr[0])
            if not sub_sub_list:
                return False
#             print(sub_list, sub_sub_list)
            sub_list += sub_sub_list
            for s in sub_list:
                rarr.remove(s)
            

    return True

# a = [1,1,2,2,2]
# a = [3,3,3,3,4]
# a = [1,1,1,1,1,2,2,3,3,3,3,3,4]
a = [5,5,5,5,4,4,4,4,3,3,3,3]
a = [13,11,1,8,6,7,8,8,6,7,8,9,8]
matchStick(a)
```
