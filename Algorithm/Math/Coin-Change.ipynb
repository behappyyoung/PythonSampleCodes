```python
"""

"""
def coinChange(coins, amount):
        dp = [amount+1] * (amount + 1)
        dp[0] = 0
        
        for coin in coins:
            print(coin)
            for x in range(coin, amount + 1):
                dp[x] = min(dp[x], dp[x - coin] + 1)
                print(coin, x, dp)
        print(dp)
        return dp[amount] if dp[amount] != (amount+1) else -1 
    

coinChange([50, 49, 48, 47,6, 4, 3], 15)
```
