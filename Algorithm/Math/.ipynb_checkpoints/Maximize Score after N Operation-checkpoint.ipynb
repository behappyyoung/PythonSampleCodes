```python
"""
    choose two elements(x, y) from list(nums)
    receive a score of i * gcd(x, y)
    remove x, y from nums
    
    nums = [3, 4, 6, 8]
    
    (1*gcd(3,6)) + (2*gcd(4,8)) = 11 
    
"""
```

```python
import math
def max_score(nums):
    
    def cal_score(nums, c_index):
        if not nums:
            return 0
        l = len(nums)
        max_score = 0
        for i in range(l):
            for j in range(i+1, l):
                x = nums[i]
                y = nums[j]
                nums_copy = nums.copy()
                nums_copy.remove(x)
                nums_copy.remove(y)
                score = c_index*math.gcd(x, y) + cal_score(nums_copy, c_index+1)
                if score> max_score:
                    max_score = score
        return max_score
    
    return cal_score(nums, 1)
```

```python
import math
def max_score_dp(nums):
    scores = {}
    def cal_score_dp(nums, c_index):
        if not nums:
            return 0
        if tuple(nums) in scores:
            return scores[tuple(nums)]
        l = len(nums)
        max_score = 0
        for i in range(l):
            for j in range(i+1, l):
                x = nums[i]
                y = nums[j]
                nums_copy = nums.copy()
                nums_copy.remove(x)
                nums_copy.remove(y)
                score = c_index*math.gcd(x, y) + cal_score_dp(nums_copy, c_index+1)
                if score> max_score:
                    max_score = score
        scores[tuple(nums)] = max_score
        print(scores)
        return max_score
    
    return cal_score_dp(nums, 1)
```

```python
nums = [3, 4, 6, 8]
print(max_score(nums))
nums = [1,2,3,4,5,6]
print(max_score(nums))

nums = [3, 4, 6, 8]
print(max_score_dp(nums))
nums = [1,2,3,4,5,6]
print(max_score_dp(nums))
```
