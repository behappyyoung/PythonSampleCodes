```python
from _LinkedLists import ListNode, print_node
"""

2 > 6 > 2 > 3 > 4 > 1 >5

head 2 -> 2 -> 1 --> 5

head_end 2 > 2 > 1 --> 5 

current None

after_head 6 -> 3 > 4 -->5 --> none

after_end 6 > 3 > 4 > 5 --> none


->1->4->3->2->5->2

head 1 > 2 -->5

head_end  1 > 2 --> 2

current 2

after_head 4 > 3 > 5 -->2

after_end 4 >3 >5 -->2


1> 1

head 
head_end

current

after_head
after_end

"""
def partition(head, x):
    if not head or not head.next:
        return head

    current = head
    head = head_end = after_head = after_end = None
    
    while current:        
                    
        if current.val >= x:
            if after_head:
                after_end.next = current
                after_end = current
            else:
                after_head = after_end = current
            
        else:
            if head: 
                head_end.next = current
                head_end = current
            else:
                head = head_end = current
                
        current = current.next
        
        print(current.val if current else 'None')
        print_node(head)
        print_node(after_head)
        
    if head:
        if after_head:
            after_end.next = None
            head_end.next = after_head
    else:
        head = after_head
    
    print_node(head)
    return head

    
lnode = ListNode(1)
lnode.next = ListNode(5)
lnode.next.next = ListNode(3)
# lnode.next.next.next = ListNode(2)
# lnode.next.next.next.next = ListNode(5)
# lnode.next.next.next.next.next = ListNode(2)


print_node(lnode)

partition(lnode, 4)

```
