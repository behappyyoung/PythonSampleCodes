```python

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

# print node
def print_node(head_node):
    c_node = head_node
    print_str = ''
    while c_node is not None:
        print_str += '->' + str(c_node.val)
        c_node = c_node.next
    print(print_str)
```

```python
"""
 quick sort

"""


def sortList(head):
        if not head or not head.next:
            return head
        pre_node = head
        c_node = pre_node.next
        swapped = True
        while swapped:
            swapped = False
            temp = head
            while c_node:
                if pre_node.val > c_node.val:
                    pre_node.next = c_node.next
                    c_node.next = pre_node
                    if head == pre_node:
                        head = c_node
                        temp = head
                    else:
                        temp.next = c_node                   
                        temp = temp.next
                        
                    c_node = pre_node.next
                    
                    swapped = True
                
                else:
                    if temp != pre_node:
                        temp = temp.next
                    pre_node = pre_node.next
                    c_node = c_node.next
                    
                    
            pre_node = head
            c_node = pre_node.next    
            
        return head
        

lnode = ListNode(6)
lnode.next = ListNode(2)
lnode.next.next = ListNode(3)
lnode.next.next.next = ListNode(4)
lnode.next.next.next.next = ListNode(7)
lnode.next.next.next.next.next = ListNode(5)
print_node(lnode)
print_node(sortList(lnode))
```

```python
"""
 using Merge Sort
 
"""

def find_middle(node):
    s = node
    if s.next:
        e = node.next
    else:
        return s, None
    while e and e.next:
        s = s.next
        e = e.next.next
    m = s.next
    s.next = None
    return node, m


        
def sorting(node):
    if not node or not node.next:
        return node
    first, middle = find_middle(node)
    sorted_left = sorting(first)
    sorted_right = sorting(middle)
    sorted_head = mergeList(sorted_left, sorted_right)
    
    return sorted_head


def mergeList(left, right):

    if not left:
        return right
    if not right:
        return left
    
    if left.val < right.val:
        sorted_node = left
        left = left.next
    else:
        sorted_node = right
        right = right.next
        
    temp = sorted_node
    
    while left and right:
        if left.val < right.val:
            temp.next = left
            left = left.next
        else:
            temp.next = right
            right = right.next
        temp = temp.next

    if not left:
        temp.next = right
    if not right:
        temp.next = left

    return sorted_node

def sortList(head):
    if not head or not head.next:
        return head
 
    sorted_head = sorting(head)
    
    return sorted_head


lnode = ListNode(6)
lnode.next = ListNode(2)
lnode.next.next = ListNode(3)
lnode.next.next.next = ListNode(4)
lnode.next.next.next.next = ListNode(7)
lnode.next.next.next.next.next = ListNode(5)

print_node(lnode)

print_node(sortList(lnode))
```
