```python
def max_subarray_product_dp(nums):
    if len(nums) < 1:
        return []
    max_p = nums[0]
    current_max = nums[0]
    current_min = nums[0]
    for i in range(1, len(nums)):
        if nums[i] == 0:
            current_max = current_min = nums[i]
        else:
            pre_max = current_max
            pre_min = current_min
        
            current_max = max(nums[i], pre_max*nums[i], pre_min*nums[i])
            current_min = min(nums[i], pre_max*nums[i], pre_min*nums[i])
        
        max_p = max(current_max, max_p)
        print(i, current_max, max_p)
    return max_p
    
test_array = [2, -1, -3,0, -2, -1, 4]
max_subarray_product_dp(test_array)
```
