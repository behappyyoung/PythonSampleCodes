```python
def knapSack_yp(total_w, wt_arr, val_arr):
    # Base Case
    lw = len(wt_arr)
    if lw == 0 or total_w == 0:
        return 0

    def khelper(subt, s):  # t : current total, s : start index
        if s >= lw:
            return 0
        t_max = 0
        for i in range(s, lw - 1):
            t = 1
            while (subt - wt_arr[i] * t) >= 0:
                without_max = khelper(subt, i + 1)
                add_more = val_arr[i] * t + khelper(subt - wt_arr[i] * t, i + 1)
                c_max = max(without_max, add_more)

#                 print(subt, wt_arr[i] * t, t_max, c_max, i, t, prev_max, add_more)
                t_max = max(t_max, c_max)
                t += 1
        return t_max

    return khelper(total_w, 0)

Cap = [1, 2, 3]
Cost = [5, 2, 3]
print(knapSack_yp(4, Cap, Cost))
```

```python
def knapSack_min_yp(tw, wt_arr, val):
    la = len(wt_arr)
    max_int = (1<<31)-1
    dp = [[max_int]*(tw+1) for _ in range(la+1)]
    wt_arr, val = zip(*sorted(zip(wt_arr, val)))
    
    for i in range(la+1):
        for j in range(tw+1):
            if i == 0 or j == 0:
                dp[i][j] = max_int
            elif wt_arr[i-1] <=j:
                prev_dp = dp[i][j-wt_arr[i-1]]
                if prev_dp != max_int:
                    dp[i][j] = min(val[i-1]+prev_dp, dp[i-1][j]) 
                else:
                    dp[i][j] = min(val[i-1], dp[i-1][j]) 
            else:
                dp[i][j] = dp[i-1][j]
            
    print(dp)

    return dp[la][tw]

Cap = [2, 3, 1]
Cost = [5, 3, 2]
print(knapSack_min_yp(7, Cap, Cost))
```
