```python
def snapdeal(A, B, C):
    
    if not A:
        return 0
    
    max_num = (1<<31) - 1
    def getdeal(cap):
        
        dp = [[max_num]*(len(B)+1) for _ in range(cap+1)]
        
        dp[0][0] = max_num
        
        for i in range(cap):
            for j in range(len(B)):
                if cap>=B[j]:
                    dp[i][j] = min(dp[i][j-1], getdeal(cap-B[j]))
                else:
                    dp[i][j] = dp[i][j-1]
        print(dp)
        return dp[cap][len(B)]
    c_sum = 0
    for i in range(len(A)):
            c_sum += getdeal(A[i])
    return c_sum
            
F = [4,5]
Cap = [1, 3]
Cost = [5, 3]
print(snapdeal(F, Cap, Cost))
```

```python
def snapdeal_test(total_w, wt_arr, val_arr):
    
    n = len(wt_arr)
    if n == 0 or total_w == 0:
        return 0

    def khelper(subt, s):  # t : current total, s : start index
        
        if s >= n:
            return 0
        t_max = 0
        
        for i in range(s, n - 1):
            t = 1
            while (subt - wt_arr[i] * t) >= 0:
                prev_max = khelper(subt, i + 1)
                add_more = val_arr[i] * t + khelper(subt - wt_arr[i] * t, i + 1)
                c_max = max(prev_max, add_more)

                print(subt, wt_arr[i] * t, t_max, c_max, i, t, prev_max, add_more)
                t_max = max(t_max, c_max)
                t += 1
        return t_max

    return khelper(total_w, 0)



F = [4,5]
Cap = [1, 2, 3]
Cost = [5, 2, 3]
print(snapdeal_test(4, Cap, Cost))
```

```python
def knapSack_yp(total_w, wt_arr, val_arr):
    # Base Case
    lw = len(wt_arr)
    if lw == 0 or total_w == 0:
        return 0

    def khelper(subt, s):  # t : current total, s : start index
        if s >= lw:
            return 0
        t_max = 0
        for i in range(s, lw - 1):
            t = 1
            while (subt - wt_arr[i] * t) >= 0:
                without_max = khelper(subt, i + 1)
                add_more = val_arr[i] * t + khelper(subt - wt_arr[i] * t, i + 1)
                c_max = max(without_max, add_more)

                print(subt, wt_arr[i] * t, t_max, c_max, i, t, prev_max, add_more)
                t_max = max(t_max, c_max)
                t += 1
        return t_max

    return khelper(total_w, 0)

Cap = [1, 2, 3]
Cost = [5, 2, 3]
print(knapSack_yp(4, Cap, Cost))
```

```python
def knapSack_min_yp(tw, wt_arr, val):
    la = len(wt_arr)
    max_int = (1<<31)-1
    dp = [[max_int]*(tw+1) for _ in range(la+1)]
    

    
    for i in range(la+1):
        for j in range(tw+1):
            if i == 0 or j == 0:
                dp[i][j] = max_int
            elif wt_arr[i-1] <=j:
                prev_dp = dp[i][j-wt_arr[i-1]]
                if prev_dp != max_int:
                    dp[i][j] = min(val[i-1]+prev_dp, dp[i-1][j]) 
                else:
                    dp[i][j] = min(val[i-1], dp[i-1][j]) 
                print(i, j,  wt_arr[i-1], val[i-1], prev_dp, dp[i][j])
            else:
                dp[i][j] = dp[i][j-1]
            
    print(dp)

    return dp[la][tw]

Cap = [2, 3, 1]
Cost = [5, 3, 2]
print(knapSack_min_yp(7, Cap, Cost))
```
