```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

    def __str__(self):
            return str(self.val)
    
    
def make_array(root):
    if not root:
        return None
    tree_array = [root.val]
    if root.left:
        tree_array.append(root.left.val)
    else:
        tree_array.append(None)
        
    if root.right:
        tree_array.append(root.right.val)
    else:
        tree_array.append(None)
    
    left = root.left
    right = root.right
#     left_left = left.left if left else None
#     left_right = left.right if left else None
#     right_left = right.left if right else None
#     right_right = right.right if right else None
    print(tree_array, left, right)
#     while left_left or left_right or right_left or right_right:
#     if left:
#         print('left',left,  left.left, left.right)
#         if left.left:
#             tree_array.append(left.left)
#         if left.right:
#             tree_array.append(left.right)
#     if right:
#         print('right',right,  right.left, right.right)
#         if right.left:
#             tree_array.append(right.left)
#         if right.right:
#             tree_array.append(right.right)
#         tree_array.append(make_array(right)) 

    
    return tree_array

def generateTrees(n):
    if n ==1:
        return [[n]]
        
    def gen_tree(start, n):
        if n < start:
            return [None]
        elif start == n:
            return [TreeNode(n)]
        
        tree_arr = []
        for i in range(start, n+1):
            left_tree = gen_tree(start, i-1)
            right_tree = gen_tree(i+1, n)
            # make tree and join
            for j in range(len(left_tree)) :
                left = left_tree[j]
                for k in range(len(right_tree)):
                    right = right_tree[k]
                    c_tree = TreeNode(i)
                    c_tree.left = left
                    c_tree.right = right
                    tree_arr.append(c_tree)
        return tree_arr
    
    return gen_tree(1, n)

```

```python
out = generateTrees(3)
# print(out)
result_array = []
for o in out:
    ma = make_array(o)
    result_array.append(ma)
print(result_array)
```
