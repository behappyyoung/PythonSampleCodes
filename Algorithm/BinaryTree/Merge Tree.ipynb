```python

def mergeTrees(t1, t2):

    if not t1 and not t2:
        return None
    
    def sumTree(lTree, rTree):
        c_val = 0
        if not lTree and not rTree:
            return None
        
        if lTree and rTree:
            c_val =+ lTree.val + rTree.val
        elif not lTree:
            c_val += rTree.val
        elif not rTree:
            c_val += lTree.val
          
        return c_val
    
    def createTree(lTree, rTree):
           
        return_tree = TreeNode(sumTree(lTree, rTree))
        
        if lTree:
            if rTree:
                return_tree.left = createTree(lTree.left, rTree.left)
                return_tree.right = createTree(lTree.right, rTree.right)
            else:
                return_tree.left = createTree(lTree.left, None)
                return_tree.right = createTree(lTree.right, None)
    
        else:
            if rTree:
                return_tree.left = createTree(None, rTree.left)
                return_tree.right = createTree(None, rTree.right)
            else:
                return None
        
        return return_tree
        
    r_tree = createTree(t1, t2)
    
        
    return r_tree

```

```python
class PointTreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        self.next = None

    def __str__(self):
        return str(self.val)
    
def print_ptree(root):
    current_level = [root]
    while current_level:
        print(' '.join('-' if node == '-' else f'{node.val}/{node.next}' for node in current_level))
        next_level = list()
        for n in current_level:
            if n == '-':
                continue
            if n.left:
                next_level.append(n.left)
            else:
                next_level.append('-')
            if n.right:
                next_level.append(n.right)
            else:
                next_level.append('-')
        current_level = next_level
    
def connect(root):

        if not root:
            return root
        top = root
        left = top.left
        c_next = top
        while left:
            left.next = c_next.right
            c_next = c_next.next
            if c_next:
                left = left.next
                left.next = c_next.left
                left = left.next
            else:
                top = top.left
                left = top.left
                c_next = top
        return root


t1 = PointTreeNode(3)
t13 = PointTreeNode(3)
t15 = PointTreeNode(5)
t17 = PointTreeNode(7)

t1.left = t13
t1.right = t15
t15.right = t17
t15.left = PointTreeNode(1)
t13.right = PointTreeNode(9)
t13.left = PointTreeNode(6)

print_ptree(t1)
r = connect(t1)
print('=======')
print_ptree(r)
```
