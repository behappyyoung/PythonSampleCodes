```python
def generate(inputn):
    def rec(n, right, comb):
        print(n, right, comb, combs)
        if right<0 or right>n:
            return
        elif n==0:
            if right==0:
                combs.append(''.join(comb))
        else:
            comb.append('(')
            rec(n-1, right+1, comb)
            comb.pop()
            comb.append(')')
            rec(n-1, right-1, comb)
            comb.pop()
    combs = []
    rec(2*inputn, 0, [])
    print(combs)
    return combs
c = generate(3)
print(c)
```
