```python
"""
A sequence of numbers is called arithmetic if it consists of at least three elements and 
if the difference between any two consecutive elements is the same.
Input: nums = [2,4,6,8,10]
Output: 7
Explanation: All arithmetic subsequence slices are:
[2,4,6]
[4,6,8]
[6,8,10]
[2,4,6,8]
[4,6,8,10]
[2,4,6,8,10]
[2,6,10]

Input: nums = [7,7,7,7,7]
Output: 16
Explanation: Any subsequence of this array is arithmetic.
[012]
[013]
[014]
[023]
[024]
[034]
[123]
[124]
[134]
[234]
[0123]
[0124]
[0134]
[0234]
[1234]
[01234]
"""
```

```python
from itertools import combinations
  

def arithmeticSub(arr):
    t_sub = {}
    
    l = len(arr)
    cc = 0
    if l < 3:
        return 0
    
    for i in range(2, l):
        c_sub = {}
        for j in range(i-1, -1, -1):
            diff = arr[i] - arr[j]     
            k = j -1
            c_d = arr[j]
            under_j = 0
            
            while k >=0:
                print('b', i, j, k)
                if diff == c_d - arr[k]:
                    under_j += 1
                    kk = k-1
                    while kk >=0:
                        if arr[k] == arr[kk]:
                            under_j +=1
                            kk -= 1
                        else:
                            kk = -1
                    
                    c_d = arr[k]
                    k -= 1
                elif diff > (c_d - arr[k]):
                    k -= 1
                else:
                    k = -1
                    
            
                print('af',i, j, k, under_j)
            if diff in c_sub:
                c_sub[diff] += under_j
            else:
                c_sub[diff] = under_j
            
            print('in j', i, j, diff,  under_j, c_sub, '/', t_sub)
            
            if diff in t_sub:
                t_sub[diff] += c_sub[diff]
            else:
                t_sub[diff] = c_sub[diff]
            
            
    
    for ts in t_sub:
        cc += t_sub[ts]
    return cc


```

```python
# nums = [7,7,7,7,7]
nums = [2, 4, 6, 8, 10]
nums = [0,1,2,2,2]
print(arithmeticSub(nums))
```
