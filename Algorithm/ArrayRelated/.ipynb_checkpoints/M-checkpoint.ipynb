```python
def countSmaller(nums):
    if not nums:
        return nums
    answers = []
    for i in range(len(nums)-1):
        count = 0
        for j in range(i+1, len(nums)):
            if nums[i] > nums[j]:
                count += 1
        answers.append(count)
    answers.append(0)
    return answers
    
nums = [21,5,2,6,20,3,1]
countSmaller(nums)
```

```python
"""
new custom TreeNode with left / same counts

"""
class TreeNode:
    def __init__(self, x, lcount=0, samecount=1):
        self.val = x
        self.left = None
        self.right = None
        self.lcount = lcount
        self.samecount = samecount
        

    def __str__(self):
        return str(self.val)

def print_btree(root):
    current_level = [root]
    while current_level:
        print(' '.join('-' if node == '-' else f'{node.val}/{node.lcount}' for node in current_level))
        next_level = list()
        for n in current_level:
            if n == '-':
                continue
            if n.left:
                next_level.append(n.left)
            else:
                next_level.append('-')
            if n.right:
                next_level.append(n.right)
            else:
                next_level.append('-')
        current_level = next_level
        
def countSmaller_bst(nums):
    if not nums:
        return nums
    
    answers = [0]
    bst = TreeNode(nums[len(nums)-1])
    
    def build_tree(c, c_root, pre_lcount=0):
        if not c_root:
            answers.insert(0, pre_lcount)
            print(c, pre_lcount, answers)
            return TreeNode(c, pre_lcount)
        
        if c == c_root.val:
            answers.insert(0, pre_lcount)
            c_root.samecount += 1
        elif c < c_root.val:
            c_root.lcount += 1
            c_root.left = build_tree(c, c_root.left, pre_lcount)
        elif c > c_root.val:
            c_root.right = build_tree(c, c_root.right, max(pre_lcount, c_root.lcount)+1*c_root.samecount)

        return c_root

    for i in range(len(nums)-2, -1, -1):
        if nums[i] == bst.val:
            bst.samecount += 1
            answers.insert(0, bst.lcount)
        else:
            build_tree(nums[i], bst)    
    
    
    print_btree(bst)
    return answers
    
# nums = [5,2,6,1]
# nums = [5183,2271,3067,539,8939,2999,9264,737,3974,5846,-210,9278,5800,2675,6608,1133,-1,6018,9672,5179,9842,7424,-209,2988,2757,5984,1107,2644,-499,7234,7539,6525,347,5718,-742,1797,5292,976,8752,8297,1312,3385,5924,2882,6091,-282,2595,96,1906,8014,7667,5895,7283,7974,-167,7068,3946,6223,189,1589,2058,9277,-302,8157,8256,5261,8067,1071,9470,2682,8197,5632,753,3179,8187,9042,8167,4657,7080,7801,5627,7917,8085,928,-892,-427,3685,4676,2431,8064,8537,343,505,4352,2108,4399,66,2086,1922,9126,9460,393,443,5689,7595,850,8493,2866,732,3738,7933,3666,2370,5804,4045,7903,8009,5387,5542,7593,6862,1547,6934,-160,9693,4560,7429,9989,7232,-594,587,6476,9277,4471,5979,6268,2419,6706,-727,1927,7361,9684,5519,2703,1723,5181,3545,4290,9421,4288,1656,1541,9632,1448,-490,4747,5416,4139,-845,3834,3349,8594,7882,2279,7777,9369,9917,8167,6799,-612,5604,5787,2615,7033,5986,-322,8631,1793,-612,3528,206,419,1413,8585,5658,-981,1391,8088,7035,6259,-651,3118,9105,4531,2569,7576,7981,838,5715,1387,8506,331,7844,9187,681]
nums = [4, 22,5,7, 21,6,20,5, 1, 5, 7]
# nums = [26,78,27,100,33,67,90,23,66,5,38,7,35,23,52,22,83,51,98,69,81,32,78,28,94,13,2,97,3,76,99,51,9,21,84,66,65,36,100,41]
countSmaller_bst(nums)
```

```python
"""
normal TreeNode

"""
class TreeNode:
    def __init__(self, x, lcount=0, samecount=0):
        self.val = x
        self.left = None
        self.right = None
        

    def __str__(self):
        return str(self.val)

def print_btree(root):
    current_level = [root]
    while current_level:
        print(' '.join('-' if node == '-' else f'{node.val}' for node in current_level))
        next_level = list()
        for n in current_level:
            if n == '-':
                continue
            if n.left:
                next_level.append(n.left)
            else:
                next_level.append('-')
            if n.right:
                next_level.append(n.right)
            else:
                next_level.append('-')
        current_level = next_level
        

def countSmaller_bst_a(nums):
    if not nums:
        return nums
    
    answers = [0]
    left_counts = dict()
    same_counts = dict()
    
    
    bst = TreeNode(nums[len(nums)-1])
    left_counts[nums[len(nums)-1]] = 0
    
    def build_tree(c, c_root, last_left, pre_lcount=0):
        if not c_root:
            answers.insert(0, pre_lcount)
            print('--', c, last_left, pre_lcount, answers, left_counts, same_counts)
            left_counts[c] = pre_lcount
            return TreeNode(c)
            
        if c == c_root.val:
            answers.insert(0, pre_lcount)
            same_counts[c] = same_counts.get(c, 1) +1
            return None
        elif c < c_root.val:
            if last_left:
                left_counts[c_root.val] = left_counts.get(last_left, 0) + same_counts.get(last_left, 1)
            else:
                left_counts[c_root.val] = 0
                
            c_root.left = build_tree(c, c_root.left, last_left)
        elif c > c_root.val:
            print(c, c_root.val, max(pre_lcount, left_counts.get(c_root.val, 0))+1)
            last_left = c_root.val
            c_root.right = build_tree(c, c_root.right, last_left, max(pre_lcount, left_counts.get(last_left, 0))+same_counts.get(last_left, 1))

        return c_root

    for i in range(len(nums)-2, -1, -1):
        last_left = None
        if nums[i] == bst.val:
            same_counts[nums[i]] = same_counts.get(nums[i], 1) +  1
            answers.insert(0, left_counts.get(nums[i], 0))
        else:
            build_tree(nums[i], bst, last_left)    
    
    print(left_counts)
    print_btree(bst)
    return answers
    
# nums = [5,2,6,1]
# nums = [5183,2271,3067,539,8939,2999,9264,737,3974,5846,-210,9278,5800,2675,6608,1133,-1,6018,9672,5179,9842,7424,-209,2988,2757,5984,1107,2644,-499,7234,7539,6525,347,5718,-742,1797,5292,976,8752,8297,1312,3385,5924,2882,6091,-282,2595,96,1906,8014,7667,5895,7283,7974,-167,7068,3946,6223,189,1589,2058,9277,-302,8157,8256,5261,8067,1071,9470,2682,8197,5632,753,3179,8187,9042,8167,4657,7080,7801,5627,7917,8085,928,-892,-427,3685,4676,2431,8064,8537,343,505,4352,2108,4399,66,2086,1922,9126,9460,393,443,5689,7595,850,8493,2866,732,3738,7933,3666,2370,5804,4045,7903,8009,5387,5542,7593,6862,1547,6934,-160,9693,4560,7429,9989,7232,-594,587,6476,9277,4471,5979,6268,2419,6706,-727,1927,7361,9684,5519,2703,1723,5181,3545,4290,9421,4288,1656,1541,9632,1448,-490,4747,5416,4139,-845,3834,3349,8594,7882,2279,7777,9369,9917,8167,6799,-612,5604,5787,2615,7033,5986,-322,8631,1793,-612,3528,206,419,1413,8585,5658,-981,1391,8088,7035,6259,-651,3118,9105,4531,2569,7576,7981,838,5715,1387,8506,331,7844,9187,681]
nums = [4, 4, 22,5,21,6, 20,7, 1, 7, 5, 7]
countSmaller_bst_a(nums)
```
