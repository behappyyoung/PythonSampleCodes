```python
"""
Input: nums = [2,0,2,1,1,0]
Output: [0,0,1,1,2,2]
"""
def sortColors(nums):

        """
        Do not return anything, modify nums in-place instead.
        """
       
        zero_count = 0
        one_count = 0
        two_count = 0
        
        for n in nums:
            if n==0:
                zero_count += 1
            elif n==1:
                one_count += 1
            else:
                two_count += 1
                
        nums.clear()
        
        for i in range(zero_count):
            nums.append(0)
        for i in range(one_count):
            nums.append(1)
        for i in range(two_count):
            nums.append(2)

        return nums
    
nums = [2,0,2,1,1,0]
sortColors(nums)
```
