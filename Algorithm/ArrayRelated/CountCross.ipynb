```python
def countCross(n, mines):
    
    plist = [[1 for i in range(n)] for j in range(n)]            
    for m in mines:
        plist[m[0]][m[1]] = 0
    
#     for p in plist:
#         print(p)
    
    left_max_list = [[plist[j][i] for i in range(n)] for j in range(n)]  
    up_max_list = [[plist[j][i] for i in range(n)] for j in range(n)]  
    
    def countLeft(r, c):
        if c == 0:
            return plist[r][0]
        p_max = left_max_list[r][c-1]
        c_max = p_max + 1
        left_max_list[r][c] = c_max
        return c_max
        
    def countUp(r, c):
        if r == 0:
            return plist[0][c]
        p_max = up_max_list[r-1][c]
        c_max = p_max + 1
        up_max_list[r][c] = c_max
        return c_max
    
    def countRight(r, c, max_c):
        if c >= n:
            return 0
        r_count = 1
        max_count = min(max_c+c, n)
        for i in range(c+1, max_count):
            if plist[r][i] == 0:
                return r_count
            r_count += 1
        return r_count
    
    def countDown(r, c, max_r):
        if r >=n:
            return 0
        d_count = 1
        max_count = min(max_r+r, n)
        for i in range(r+1, max_count):
            if plist[i][c] == 0:
                return d_count
            d_count += 1
            
        return d_count
    
    
    max_cross = 0
    for r in range(n):
        for c in range(n):
            if plist[r][c] != 0:
                l_max = countLeft(r, c)
                u_max = countUp(r, c)
                r_max = countRight(r, c, min(l_max, u_max)+1)
                d_max = countDown(r, c, min(l_max, u_max)+1)
                c_min = min(r_max, d_max, l_max, u_max)
#                 print(r, c, r_max, d_max, l_max, u_max)
                max_cross = max(max_cross, c_min)
                plist[r][c] = c_min

#     print('-----------------------')
#     for p in left_max_list:
#         print(p)
#     for p in up_max_list:
#         print(p)
    
#     print('-----------------------')
#     for p in plist:
#         print(p)
    return max_cross
```
